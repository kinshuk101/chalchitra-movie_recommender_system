```python
import numpy as np
import pandas as pd
```

```python
movies = pd.read_csv("tmdb_5000_movies.csv")
credits= pd.read_csv("tmdb_5000_credits.csv")
```

```python
movies.head()
```

```python
credits.head(1)['crew']
```

```python
movies=movies.merge(credits,on="title")
```

```python
movies
```

```python
#task of tag creation
#genres, id, keywords, title, overview,cast, crew
```

```python
movies.info()
```

```python
movies = movies[['id','title','overview','genres','keywords','cast','crew']] 
```

```python
#removing null items
movies.isnull().sum()
```

```python
movies.dropna(inplace=True)
```

```python
movies.iloc[0].genres
```

```python
import ast
```

```python
def convert(obj):
    L=[]
    for i in ast.literal_eval(obj):
        L.append(i['name'])
    return L    
```

```python
movies['genres'] = movies['genres'].apply(convert)
```

```python
movies['keywords'] = movies['keywords'].apply(convert)
```

```python
movies.iloc[0].keywords
```

```python
def extract(obj):
    L=[]
    counter=0
    for i in obj:
        if(counter<3):
            L.append(i)
            counter+=1
        else:
            break
    return L
```

```python
movies['cast']=movies['cast'].apply(convert)
```

```python
movies['cast']=movies['cast'].apply(extract)
```

```python
movies.head()
```

```python
movies['cast'].head()
```

```python
movies['crew'][0]
```

```python
def convert2(obj):
    L=[]
    for i in ast.literal_eval(obj):
        if i["job"]== "Director":
            L.append(i["name"])
            break
    return L    
```

```python
movies['crew'] = movies['crew'].apply(convert2)
```

```python
# movies['overview'][4437] = 'the me'
# movies['overview'][2662] = 'the me'
# movies['overview'][4147] = 'the me'
movies['overview']=movies['overview'].apply(lambda x:x.split())
# for i in range(4809):
#     movies['overview'][i] = movies['overview'][i].split()#list to string
```

```python
movies['genres'] = movies['genres'].apply(lambda x:[i.replace(" ","") for i in x])#removing spaces
movies['keywords'] = movies['keywords'].apply(lambda x:[i.replace(" ","") for i in x])
movies['cast'] = movies['cast'].apply(lambda x:[i.replace(" ","") for i in x])
movies['crew'] = movies['crew'].apply(lambda x:[i.replace(" ","") for i in x])
```

```python
movies.head()
```

```python
movies['tags']= movies['overview'] + movies['genres'] + movies['keywords']+movies['cast']+movies['crew']
```

```python
new_df= movies[['id','title','tags']]
```

```python
new_df
```

```python
new_df['tags'] = new_df['tags'].apply(lambda x:" ".join(x)) #list to stirng
```

```python
new_df['tags'][0]
```

```python
new_df['tags']=new_df['tags'].apply(lambda x:x.lower())
```

```python
import nltk
```

```python
from nltk.stem.porter import PorterStemmer
ps= PorterStemmer()
```

```python
def stem(obj):
    L=[]
    for i in obj.split():
        L.append(ps.stem(i))
        
    return " ".join(L)
```

```python
new_df['tags'] = new_df['tags'].apply(stem)
```

```python
from sklearn.feature_extraction.text import CountVectorizer # to remove stop words and 5000 words to be taken
cv=CountVectorizer(max_features=2000, stop_words='english')
```

```python
vectors = cv.fit_transform(new_df['tags']).toarray()
#vector bn chuke 4800 vectors hai or 5000 words h jo jo unme aa rhe h unki freq k hisab se
```

```python
cv.get_feature_names() #5000 frequent words
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(vectors)
```

```python
similarity[0]
```

```python
def recommend(movie):
    movie_index=new_df[new_df['title']==movie].index[0]  # isse movie ka index milega in data frame
    distances=similarity[movie_index]
    movie_list = sorted(list(enumerate(distances)),reverse=True,key=lambda x:x[1])[1:6]
    
    for i in movie_list:
        ans=i[0]
        print (new_df.iloc[ans].title)
```

```python
sorted(list(enumerate(similarity[0])),reverse=True,key=lambda x:x[1])[1:6]  #5 movies
```

```python
recommend('Avatar')
```
